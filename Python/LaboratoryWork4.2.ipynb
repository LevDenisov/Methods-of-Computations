## Алгоритмы приближенного вычисления интеграла по квадратурным формулам 

#### CКФ левого прямоугольника

```python
def complex_left_rectangle_method(f, a, b, m):
    h = (b - a) / m
    return h * sum(f(a + i * h) for i in range(0, m))
```

#### CКФ правого прямоугольника

```python
def complex_right_rectangle_method(f, a, b, m):
    h = (b - a) / m
    return h * sum(f(a + i * h) for i in range(1, m + 1))
```

#### CКФ среднего прямоугольника

```python
def complex_middle_rectangle_method(f, a, b, m):
    h = (b - a) / m
    return h * sum(f(a + i * h + h / 2) for i in range(0, m))
```

#### СКФ трапеций

```python
def complex_trapeze_method(f, a, b, m):
    return (complex_left_rectangle_method(f, a, b, m) + complex_right_rectangle_method(f, a, b, m)) / 2
```

#### СКФ Симпсона

```python
def complex_simpson_method(f, a, b, m):
    return (complex_left_rectangle_method(f, a, b, m) + complex_right_rectangle_method(f, a, b, m) +
            4 * complex_middle_rectangle_method(f, a, b, m)) / 6
```

## Произвольные функции

#### Произвольная функция №1

```python
def f_custom(x):
    return math.cos(x) - 2 * x


def der_f_custom(x):
    return (-1) * math.sin(x) - 2


def integral_f_custom(a, b):
    return math.sin(b) - math.sin(a) + a * a - b * b
```

```python
def f_3(x):
    return 4 * (x ** 3)


def der_f_3(x):
    return 12 * (x ** 2)


def der_2_f_3(x):
    return 24 * x


def der_3_f_3(x):
    return 24


def der_4_f_3(x):
    return 0


def integral_f_3(a, b):
    return b ** 4 - a ** 4
```
